# Cronbachs Alpha

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

def cronbach_alpha(df):
    k = df.shape[1]
    item_var = df.var(axis=0, ddof=1)
    total_var = df.sum(axis=1).var(ddof=1)
    return (k / (k - 1)) * (1 - (item_var.sum() / total_var))

def item_total_correlation(df):
    corrs = {}
    total_score = df.sum(axis=1)
    for col in df.columns:
        item_score = df[col]
        corrected_total = total_score - item_score
        corrs[col] = pearsonr(item_score, corrected_total)[0]
    return pd.Series(corrs)
```

```python
for i, group in enumerate([group1, group2, group3, group4, group5, group6], start=1):
    print(f"Factor {i}:")
    print("Cronbach's Alpha:", round(cronbach_alpha(group.dropna()), 3))
    print("Item-Total Correlations:")
    print(item_total_correlation(group.dropna()).round(3))
    print("-" * 40)
```

# CR and AVE for M2

```python
def compute_cr_ave_semopy(model, data):
    import numpy as np
    import pandas as pd

    # Get unstandardized loadings
    estimates = model.inspect()
    factors = estimates[estimates["op"] == "~"]["rval"].unique()

    result = {}

    for factor in factors:
        # Get observed variables linked to this factor
        loadings = estimates[(estimates["op"] == "~") & (estimates["rval"] == factor)]
        items = loadings["lval"].values
        lambdas = []

        # Manually standardize loadings
        for i, row in loadings.iterrows():
            item = row["lval"]
            unstd_loading = row["Estimate"]
            item_std = data[item].std()
            factor_var = estimates[(estimates["lval"] == factor) & (estimates["rval"] == factor) & (estimates["op"] == "~~")]["Estimate"].values[0]
            standardized_loading = unstd_loading * (factor_var ** 0.5) / item_std
            lambdas.append(standardized_loading)

        lambdas = np.array(lambdas)
        error_vars = 1 - lambdas**2

        # CR and AVE
        cr = np.sum(lambdas**2) / (np.sum(lambdas**2) + np.sum(error_vars))
        ave = np.sum(lambdas**2) / len(lambdas)

        result[factor] = {'CR': round(cr, 3), 'AVE': round(ave, 3)}

    return pd.DataFrame(result).T
```

```python
import pandas as pd
import numpy as np
from itertools import combinations

# Define item groups for each latent factor in M2
latent_groups_m2 = {
    'Factor1': ['PV1', 'PV2', 'PV3', 'PV4', 'PV5', 'PD1', 'PD2', 'PD3', 'PD4', 'PD5'],
    'Factor2': ['Loan1', 'Loan2', 'Loan4', 'Resale1', 'Resale2', 'Resale3'],
    'Factor3': ['Price1', 'Price2', 'Price3'],
    'Factor4': ['Safety1', 'Safety2', 'Safety3'],
    'Factor5': ['Brand1', 'Brand2', 'Brand3'],
    'Factor6': ['Service1', 'Service2', 'Service3']
}

# Function to compute HTMT
def compute_htmt_from_groups(data, latent_groups):
    htmt_matrix = pd.DataFrame(index=latent_groups.keys(), columns=latent_groups.keys(), dtype=float)

    for f1, f2 in combinations(latent_groups.keys(), 2):
        items_f1 = latent_groups[f1]
        items_f2 = latent_groups[f2]

        # Heterotrait-heteromethod correlations (between constructs)
        hetero_corrs = [abs(data[i1].corr(data[i2])) for i1 in items_f1 for i2 in items_f2]

        # Monotrait-heteromethod correlations (within each construct)
        mono_f1 = [abs(data[i1].corr(data[i2])) for i1, i2 in combinations(items_f1, 2)]
        mono_f2 = [abs(data[i1].corr(data[i2])) for i1, i2 in combinations(items_f2, 2)]

        htmt_val = np.mean(hetero_corrs) / np.sqrt(np.mean(mono_f1) * np.mean(mono_f2))
        htmt_matrix.loc[f1, f2] = htmt_val
        htmt_matrix.loc[f2, f1] = htmt_val

    # Diagonal = 1
    np.fill_diagonal(htmt_matrix.values, 1.0)

    return htmt_matrix

# Run HTMT for M2
htmt_m2 = compute_htmt_from_groups(df, latent_groups_m2)
print("── HTMT Matrix (Model 2) ───────────────────────")
print(htmt_m2.round(3))
```

```python
def compute_cr_ave_semopy(model, data):
    import numpy as np
    import pandas as pd

    # Extract estimates
    estimates = model.inspect()

    # Get list of latent variables (factors)
    latent_factors = estimates[estimates["op"] == "~"]["rval"].unique()

    result = {}

    for factor in latent_factors:
        # Get loadings (item ~ factor)
        loadings = estimates[(estimates["op"] == "~") & (estimates["rval"] == factor)]
        lambdas = []

        # Get factor variance (from the diagonal of latent covariances)
        factor_var = estimates[
            (estimates["lval"] == factor) &
            (estimates["rval"] == factor) &
            (estimates["op"] == "~~")
        ]["Estimate"].values[0]

        for _, row in loadings.iterrows():
            item = row["lval"]
            unstd_loading = row["Estimate"]
            item_std = data[item].std()

            # Standardized loading formula
            std_loading = unstd_loading * (factor_var ** 0.5) / item_std
            lambdas.append(std_loading)

        lambdas = np.array(lambdas)
        error_vars = 1 - lambdas**2

        # Calculate CR and AVE
        cr = np.sum(lambdas**2) / (np.sum(lambdas**2) + np.sum(error_vars))
        ave = np.sum(lambdas**2) / len(lambdas)

        result[factor] = {'CR': round(cr, 3), 'AVE': round(ave, 3)}

    return pd.DataFrame(result).T
```

```python
import numpy as np
import pandas as pd
from itertools import combinations

# Define M1 latent constructs and their items
m1_constructs = {
    'Factor3': ['Loan1', 'Loan2', 'Loan4', 'Resale1', 'Resale2', 'Resale3'],
    'Factor4': ['Price1', 'Price2', 'Price3'],
    'Factor5': ['Safety1', 'Safety2', 'Safety3'],
    'Factor6': ['Brand1', 'Brand2', 'Brand3'],
    'Factor7': ['Service1', 'Service2', 'Service3'],
    'Factor1': ['PV1', 'PV2', 'PV3', 'PV4', 'PV5'],
    'Factor2': ['PD1', 'PD2', 'PD3', 'PD4', 'PD5']
}

def compute_htmt(df, constructs):
    htmt_matrix = pd.DataFrame(index=constructs.keys(), columns=constructs.keys(), dtype=float)

    for c1, c2 in combinations(constructs.keys(), 2):
        items_c1 = constructs[c1]
        items_c2 = constructs[c2]

        # Compute all pairwise correlations between items from different constructs
        corr_values = []
        for i in items_c1:
            for j in items_c2:
                corr = df[i].corr(df[j])
                corr_values.append(abs(corr))
        htmt_value = np.mean(corr_values)

        htmt_matrix.loc[c1, c2] = htmt_value
        htmt_matrix.loc[c2, c1] = htmt_value

    # Diagonal should be 1
    for c in constructs.keys():
        htmt_matrix.loc[c, c] = 1.0

    return htmt_matrix.round(3)

# Run HTMT on the full dataframe (or dropna if needed)
htmt_m1 = compute_htmt(df.dropna(), m1_constructs)

print("\n── HTMT Matrix (Model 1) ───────────────────────")
print(htmt_m1)
```
